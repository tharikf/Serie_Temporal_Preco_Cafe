#### Avaliando em Teste

```python
modeloV1_MM_teste = teste_diferenciado[['preco_reais']]
modeloV1_MM_teste.columns = ['Observado']

historico = [modeloV1_MM_treino.iloc[i, 0] for i in range(len(modeloV1_MM_treino))]
lista_previsao = []

for i in range(len(modeloV1_MM_teste)):
    previsao = np.mean(historico[-5:])
    observacao = modeloV1_MM_teste.iloc[i, 0]
    lista_previsao.append(previsao)
    historico.append(observacao)

modeloV1_MM_teste['Previsto'] = lista_previsao
```

```python
avaliacao_teste = AvaliandoModelos(modeloV1_MM_teste['Observado'],
                                   modeloV1_MM_teste['Previsto'],
                                   nome_col = 'ModeloV1_MM_5Dias', nome_indice = 'Teste')
```

```python
avaliacao_teste
```

```python
plt.figure(figsize = (8, 6))
plt.plot(modeloV1_MM_teste['Observado'])
plt.plot(modeloV1_MM_teste['Previsto'])
plt.title('Preço da saca de 60kg de soja - Preco Original x Preco Previsto')
plt.ylabel('')
plt.show();
```

```python
residuosV1_teste = PlotResiduos(modeloV1_MM_teste)
residuosV1_teste.plot_residuos()
```

#### Avaliando em treino

#### Avaliando em teste

```python
modeloV2_MME_teste = teste_diferenciado[['preco_reais']]
modeloV2_MME_teste.columns = ['Observado']

hist = [modeloV2_MME_treino.iloc[i, 0] for i in range(len(modeloV2_MME_treino))]
hist_prev = [modeloV2_MME_treino.iloc[i, 1] for i in range(len(modeloV2_MME_treino))]
lista_previsao = []

for t in range(len(modeloV2_MME_teste)):
    yhat = hist_prev[-1] + (alpha * (hist[-1] - hist_prev[-1]))
    obs = modeloV2_MME_teste.iloc[t, 0]
    lista_previsao.append(yhat)
    hist.append(obs)
    hist_prev.append(yhat)

modeloV2_MME_teste['Previsto'] = lista_previsao
```

```python
avaliacao_teste['ModeloV2_MME'] = AvaliandoModelos(modeloV2_avalia_teste['Preco_Original'],
                                                   modeloV2_avalia_teste['Preco_Previsto'])
```

#### Interpretando coeficientes

Ljung-Box (Q): Testa se os resíduos são independentes, ou seja, não possuem autocorrelação.\
* Hipótese Nula: Resíduos independentes.
* Hipótese Alterantiva: Resíduos possuem autocorrelação.

Heteroskedasticity (H): Testa se os resíduos são homocedásticos.
* Hipótese Nula: Resíduos são homocedásticos.
* Hipótese Alternativa: Resíduos são heterocedásticos.

Jarque-Bera(JB): Testa se os resíduos são normalmente distribuídos.
* Hipótese Nula: Resíduos são normalmente distribuídos.
* Hipótese Alternativa: Resíduos não são normalmente distribuídos.

```python
modelo_v4.summary()
```

#### Avaliando em teste

```python
modeloV4_ARMA_teste = teste_diferenciado[['preco_reais']]
modeloV4_ARMA_teste.columns = ['Observado']

coef_ar1 = modelo_v4.arparams.item()
coer_ma1 = modelo_v4.maparams.item()

hist = [modeloV4_ARMA_treino.iloc[i, 0] for i in range(len(modeloV4_ARMA_treino))]
error = [0.0 for i in range(len(modeloV4_ARMA_treino))]

lista_previsao = []

for t in range(len(modeloV4_ARMA_teste)):
    yhat = ((hist[-1] * coef_ar1) + (error[-1] * coer_ma1))
    obs = modeloV4_ARMA_teste.iloc[t, 0]
    lista_previsao.append(yhat)
    hist.append(obs)
    error.append(obs - yhat)

modeloV4_ARMA_teste['Previsto'] = lista_previsao
```

```python
avaliacao_teste['ModeloV4_ARMA11'] = AvaliandoModelos(modeloV4_avalia_teste['Preco_Original'],
                                                      modeloV4_avalia_teste['Preco_Previsto'])
```

```python
avaliacao_teste
```

```python
plt.figure(figsize = (8, 6))
plt.plot(modeloV4_avalia_teste['Preco_Original'])
plt.plot(modeloV4_avalia_teste['Preco_Previsto'])
plt.title('Preço da saca de 60kg de soja - Preco Original x Preco Previsto')
plt.ylabel('')
plt.show();
```

```python
residuosV4_teste = PlotResiduos(modeloV4_avalia_teste)
residuosV4_teste.plot_residuos()
```
